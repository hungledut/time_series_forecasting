```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
def get_Table(month,year,day):
    day_ = str(year) + str(month) + str(day)
    url = f"https://www.timeanddate.com/weather/vietnam/da-nang/historic?month={month}&year={year}&hd={day_}"
    print(url)
    page = requests.get(url)
    soup = BeautifulSoup(page.content, "html.parser")
    table = soup.find('table' ,id="wt-his")
    return table
```

```python
def Crawl(years,months,sample_size):
    #Tạo dataframe
    headers = ['Time','Temp','Weather','Wind','Humidity','Barometer','Visibility','Day','Month','Year']
    mydata = pd.DataFrame(columns = headers)
    months = [str(i).zfill(2) for i in months]
    days = []
    Count = 0
    for year in years:
        if Count > sample_size:
            break
        for month in months:
            if Count > sample_size:
                break
            if month in ['01','03','05','07','08','10','12']:
                days = [str(i).zfill(2) for i in range(1,32)] # 31 ngày
            if month in ['04','06','09','11']:
                days = [str(i).zfill(2) for i in range(1,31)] # 30 ngày
            if month == '02' and int(year)%4 == 0:
                days = [str(i).zfill(2) for i in range(1,30)] # 29 ngày
            if  month == '02' and int(year)%4 != 0:
                days = [str(i).zfill(2) for i in range(1,30)] # 28 ngày
            for day in days:
                if Count > sample_size:
                    break
                table = get_Table(month,year,day)
                #Lấy thời gian
                times = []
                others = []
                for j in table.find_all('tr')[2:-1] :
                    if Count > sample_size:
                        break
                    time_data = j.find_all('th')
                    time = [i.text for i in time_data]
                    x = time[0][0:8]
                    time_hour = []
                    time_hour.append(x)
                    times.append(time_hour)
                    others_data = j.find_all('td')
                    other = [i.text for i in others_data]
                    others.append(other[1:4]+other[5:8]+[day,month,year])
                    Count = Count + 1
                rows = []
                for i,j in zip(times,others):
                    rows.append(i+j) 
                for row in rows:
                    length = len(mydata)
                    mydata.loc[length] = row
                
    return mydata
```

```python
years = ['2022']
months = ['3','4','5','6','7','8','9','10']
```

## Demo với sample_size = 10

```python
sample_size = 10
```

```python
Dataframe_demo = Crawl(years,months,sample_size)
```
